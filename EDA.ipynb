```python
Y = df.groupby('City/District')['Waste Generated (Tons/Day)'].sum()

plt.figure(figsize=(12,6))
Y.plot(kind='bar')
plt.title("Total Waste Generated per City")
plt.xlabel("City/District")
plt.ylabel("Waste Generated (Tons/Day)")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
waste_per_city = df.groupby('City/District')['Waste Generated (Tons/Day)'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(x='City/District', y='Waste Generated (Tons/Day)', data=waste_per_city, palette="viridis")
plt.title("Total Waste Generated per City")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10,12))
sns.barplot(y='City/District', x='Waste Generated (Tons/Day)', 
            data=waste_per_city.sort_values(by='Waste Generated (Tons/Day)', ascending=False),
            palette="mako")
plt.title("Total Waste Generated per City")
plt.show()
```

```python
df[['Landfill_Lat', 'Landfill_Long']] = df['Landfill Location (Lat, Long)'] \
    .str.split(',', expand=True).astype(float)
df.drop('Landfill Location (Lat, Long)', axis=1, inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder


categorical_cols = ['City/District', 'Waste Type', 'Disposal Method']

encoder = OneHotEncoder(drop=None, sparse_output=False, handle_unknown='ignore')


encoded = encoder.fit_transform(df[categorical_cols])
encoded_df = pd.DataFrame(encoded, columns= encoder.get_feature_names_out(categorical_cols))
df_encoded = pd.concat([df.drop(columns=categorical_cols).reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)

print(df_encoded.head())
```

```python
df_encoded
```

```python
X = df_encoded.drop(columns=['Recycling Rate (%)'])
y = df_encoded['Recycling Rate (%)']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_test, y_pred=y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

model = RandomForestRegressor(n_estimators=300, random_state=42)
model.fit(X_train, y_train)
preds = model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, preds))
mae = mean_absolute_error(y_test, preds)
r2 = r2_score(y_test, preds)

print(f"RMSE: {rmse}, MAE: {mae}, R²: {r2}")
```
