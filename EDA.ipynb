```python
# Univariate Analysis

# Distribution of Waste Generated (Tons/Day) across all cities.

# Recycling Rate (%) distribution → mean, median, skew.

# Municipal Efficiency Score (1–10) → how many cities score low vs high.

# Cost of Waste Management per ton → range, outliers.

# Awareness Campaigns Count → frequency distribution.

# 👉 Plots: histograms, boxplots, KDE plots.
```

```python
Y = df.groupby('City/District')['Waste Generated (Tons/Day)'].sum()

plt.figure(figsize=(12,6))
Y.plot(kind='bar')
plt.title("Total Waste Generated per City")
plt.xlabel("City/District")
plt.ylabel("Waste Generated (Tons/Day)")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
waste_per_city = df.groupby('City/District')['Waste Generated (Tons/Day)'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(x='City/District', y='Waste Generated (Tons/Day)', data=waste_per_city, palette="viridis")
plt.title("Total Waste Generated per City")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10,12))
sns.barplot(y='City/District', x='Waste Generated (Tons/Day)', 
            data=waste_per_city.sort_values(by='Waste Generated (Tons/Day)', ascending=False),
            palette="mako")
plt.title("Total Waste Generated per City")
plt.show()
```
